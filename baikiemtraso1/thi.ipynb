```python
import pandas as pd

names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ivan", "Jack", "Kelly", "Liam", "Mona", "Nina", "Oscar"]
ages = [25, 30, 35, 28, 22, 45, 34, 31, 27, 29, 33, 40, 26, 32, 36]
salary = [50000, 60000, 70000, 55000, 52000, 80000, 72000, 68000, 61000, 59000, 63000, 77000, 53000, 66000, 75000]

data = {
    'Name': names,
    'Age': ages,
    'Salary': salary
}

df = pd.DataFrame(data)

print(df)
```

```python
filtered_df = df[df['Age'] > 28]
print(filtered_df)
```

```python
df2 = df['Salary'].mean()
print("Giá trị trung bình của cột Salary là:", df2)
```

```python
#câu 5
group_df = df.groupby('Age')['Salary'].sum().reset_index()
print(group_df)
```

```python
#câu 6
df3_df = df.sort_values(by='Salary', ascending=False)
print(df3_df)
```

```python
#Câu 11

print("Có giá trị NaN trong từng cột:")
print(df.isna().any())

print("\nToàn bộ DataFrame có giá trị NaN:")
print(df.isna().values.any())
```

```python
#câu 12
df_age = df['Age'].mean()
df.loc[df['Age'] > 30, 'Age'] = df_age

print(df)
```

```python
#câu 13

min_age = df['Age'].min()
max_age = df['Age'].max()

df['Normalized_Age'] = (df['Age'] - min_age) / (max_age - min_age)


print(df)
```

```python
#Câu 14
def age_group(age):
    if age < 30:
        return 'Young'
    elif age < 60:
        return 'Middle aged'
    else:
        return 'Old'


df['Age_Group'] = df['Age'].apply(age_group)
print(df)
```

```python
#câu 15
df['Salary_Pct_Change'] = df['Salary'].pct_change() * 100
```
